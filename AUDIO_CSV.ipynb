```python
import csv

file_names = [
    '10 sec/Apple 01 slow.wav',
    '10 sec/Apple 02 slow.wav',
    '10 sec/Apple 03 slow.wav',
    '10 sec/Apple 04 slow.wav',
    '10 sec/Apple 05 slow.wav',
    '10 sec/Apple 06 slow.wav',
    '10 sec/Apple 07 slow.wav',
    '10 sec/Apple 08 slow.wav',
    '10 sec/Apple 09 slow.wav',
    '10 sec/Apple 10 slow.wav',
    '10 sec/Apple 01 fast.wav',
    '10 sec/Apple 02 fast.wav',
    '10 sec/Apple 03 fast.wav',
    '10 sec/Apple 04 fast.wav',
    '10 sec/Apple 05 fast.wav',
    '10 sec/Apple 06 fast.wav',
    '10 sec/Apple 07 fast.wav',
    '10 sec/Apple 08 fast.wav',
    '10 sec/Apple 09 fast.wav',
    '10 sec/Apple 10 fast.wav'
]

csv_file_path = 'apple_audio.csv'

# Write the data to the CSV file
with open(csv_file_path, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    
    # Write the header
    csv_writer.writerow(['file_name'])
    
    # Write the file names to the column
    for file_name in file_names:
        csv_writer.writerow([file_name])

print(f"CSV file '{csv_file_path}' has been created.")
```
